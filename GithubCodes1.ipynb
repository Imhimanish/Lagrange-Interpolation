## **Lagrange Interpolation**

Using Lagrangeâ€™s interpolation formula find y(10) from the following table:

2023-04-23.png

```python
x=[5,6,9,11]
y=[12,13,14,16]
xp=10 # Interpolation at this point
def Lagrange_intp(x,y,xp):
  s=0
  for i in range(len(x)):
    for j in range(len(x)):
      if i != j:
        L=(xp-x[j])/(x[i]-x[j])
 
    s += L*y[i]
 
  return s
print ('Interpolated value =' ,Lagrange_intp(x,y,xp))
```

```python
import numpy as np
import matplotlib.pyplot as plt
x=[0, 1, 2, 3]
y=[0, 1, -4, 9]
def Lagrange_intp(x,y,xp):
  s=0
  for i in range(len(x)):
    L = 1
    for j in range(len(x)):
      if i != j:
        L = L * (xp-x[j])/(x[i]-x[j])
 
      s += L*y[i]
  return s
xp = np.linspace(0, 3, 10)
yint = Lagrange_intp(x, y, xp)

print('Interpolated values =' ,yint)
plt.plot(x,y,'o')
plt.plot(xp, yint)
plt.grid()
plt.show()
```
